```python
def check(strToFind):
    with open(csvPath) as f:
        content = f.readlines()
    # you may also want to remove whitespace characters like `\n` at the end of each line
    #content = [x.strip() for x in content] # to ignor the \n
    indx=0
    a=np.array([])
    for line in content:
        indx+=1
        if line.split(';').count(strToFind)==1:
            a=np.append(a,indx)
            #print(a)
    return np.uint16(a) 
```
